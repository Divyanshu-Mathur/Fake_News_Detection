```python
import numpy as np
import pandas as pd
```

```python
true_news = pd.read_csv('True.csv')
fake_news = pd.read_csv('Fake.csv')
```

```python
true_news.head()
```

```python
fake_news.head()
```

```python
true_news['label']=1
fake_news['label']=0
```

```python
true_news.shape,fake_news.shape
```

```python
df = pd.concat([true_news, fake_news], ignore_index=True)
df
```

```python
df = df.sample(frac=1,random_state=40).reset_index(drop=True)
df
```

```python
df.drop(['title','subject','date'],axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english'))
def preprocess(text):
    text = text.lower() #lowering the text
    text = re.sub(r'[^a-z0-9\s]', '', text) #removing special characters
    text = re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , text) #removing hhtps
    tokens = word_tokenize(text)
    text = " ".join([word for word in tokens if word not in stop_words])
    return text


    
```

```python
df['text'] = df['text'].apply(preprocess)
```

```python
x=df['text']
y=df['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(max_features=5000, stop_words='english')
x_train = vec.fit_transform(x_train)
x_test = vec.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import pickle
with (open('model.pkl','wb')) as file:
    pickle.dump(model,file)
with (open('vectorizer.pkl','wb')) as file:
    pickle.dump(vec,file)
```
